```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("InterestRates_Macro.xlsx")

# Convert Quarter (Q1, Q2...)
df["Quarter_Num"] = df["Quarter"].str.replace("Q", "").astype(int)

# Create a Date column (quarterly)
df["Date"] = pd.PeriodIndex(
    year=df["Year"],
    quarter=df["Quarter_Num"],
    freq="Q"
).to_timestamp()

df.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df["Date"], df["BankOfCanada"], label="Bank of Canada Policy Rate")

plt.title("Bank of Canada Policy Rate Over Time (1980–2025)")
plt.xlabel("Year")
plt.ylabel("Interest Rate (%)")
plt.legend()
plt.grid(True)

plt.show()
```
